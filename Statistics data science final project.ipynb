Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

α = 0.05

Null Hypothesis(H0): Both average MEDV are the same

Alternative Hypothesis(H1): Both average MEDV are NOT the same

```python
df["CHAS"].value_counts()
```

```python
x=df[df['CHAS']==0]["MEDV"]
x
```

```python
y=df[df["CHAS"]==1]["MEDV"]
y
```

```python
scipy.stats.ttest_ind(x,y,axis=0,equal_var= True)
```

Conclusion: Since p-value more than alpha value of 0.05, we failed to reject null hypothesis,there no statistical significance

Q: Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

Hypohesis:
H0= three population means are equal
H1= atleast two population mean are different

```python
df["AGE"].value_counts()
```

```python
df.loc[(df["AGE"] <= 35),'age_group'] = '35 years and younger'
df.loc[(df["AGE"] > 35) & (df["AGE"]<70),'age_group'] = 'between 35 and 70 years'
df.loc[(df["AGE"] >= 70),'age_group'] = '70 years and older'
```

```python
low = df[df["age_group"] == '35 years and younger']["MEDV"]
mid = df[df["age_group"] == 'between 35 and 70 years']["MEDV"]
high = df[df["age_group"] == '70 years and older']["MEDV"]
```

```python
f_stats, p_value = scipy.stats.f_oneway(low,mid,high,axis=0)
```

```python
print("F-Statistic={0}, P-value={1}".format(f_stats,p_value))
```

conclusion: P-value is less than alpha i.e., so we reject null hypothesis and accept alternative hypothesis as there is significance difference between population means

Q: Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Hypothesis:
    H0: There is no relationship between NOX and INDUS
    H1: There is a relationship between NOX and INDUS

```python
ax = sns.scatterplot(x="NOX", y="INDUS", data=df)
```

```python
scipy.stats.pearsonr(df['NOX'], df['INDUS'])
```

Since p-value is less than 0.05 we reject null hypothesis and go with alternate hypothesis i.e., there is relationship between both NOX and INDUS

Q: What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
H0: weighted distances to five Boston employment centres are not related to median value
H1: weighted distances to five Boston employment centres are related to median value
```

```python
import statsmodels.api as sm
```

```python
X=df['DIS']
y=df['MEDV']
X=sm.add_constant(X)
model=sm.OLS(y,X).fit()
prediction=model.predict(X)
model.summary()
```

```python
Conclusion: p<0.05 so we reject null hypothesis, weighted distances to five Boston employment centres are related to median value
```
